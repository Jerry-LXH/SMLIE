```python
import pandas as pd
data = pd.read_csv('/Volumes/SMILeSSD/Optics/Ti2Tests/Laser/I-P_measurement.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.special import erf
I = np.array(data['I'])
P = np.array(data['P'])

a, b = np.polyfit(I, P, 1)
P_fit = a*I + b

plt.scatter(I, P, label="I-P data")
plt.plot(I, P_fit, label=f"fit: P = {a:.3f}*I + {b:.3f}")
plt.xlabel("I(A)")          # 横轴名称
plt.ylabel("P(mW)")          # 纵轴名称
plt.xlim(0, 2.4) 
plt.legend()
plt.show()
```
