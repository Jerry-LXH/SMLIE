### 识别系统

```python
# 检查"描述"列每个元素是否包含词典中的关键词
def identify_component(text, your_dict):
    if pd.isna(text):
        return None
    # 先排除掉 '车门'
    keys_except_chamen = [key for key in your_dict if key != '车门']
    # 长词优先
    keys_except_chamen = sorted(keys_except_chamen, key=lambda x: len(x), reverse=True)

    for key in keys_except_chamen:
        if key in text:
            return key
    # 最后再匹配 '车门'
    if '门' in text:
        return '车门'
    return None


def identify_device(text, your_dict):
    if pd.isna(text) or str(text).strip() == '':
        return None  # 处理空值或空字符串

    text = str(text)  # 保证 text 是字符串
    for item in your_dict:
        if item == '无法打开':
            if '未开启' in text or '无法打开' in text:
                return item
        elif item == '无法关闭':
            if '未关闭' in text or '关不上' in text or '无法关闭' in text or '无法锁闭' in text:
                return item
        elif item == '松动':
            if '松动' in text or '松' in text:
                return item
        elif item == '音轻':
            if '音轻' in text or '不响' in text:
                return item
        else:
            if item in text:
                return item  # 返回匹配到的设备
    return None  # 没有匹配到

def identify_solution(text, solution_dict):
    if pd.isna(text):
        return None

    # 定义需要匹配的关键词列表
    solution_keys = sorted(solution_dict, key=lambda x: len(x), reverse=True)

    for key in solution_keys:
        # 检查每一个字是否都在文本中出现
        if all(char in text for char in key):
            return key

    return None

# 应用到新表的某一列上
data_result = pd.DataFrame()
data_result['型号'] = data_new['车型-列车号']
data_result['故障情况'] = data_new['故障情况'].apply(lambda x: identify_component(x, comp_dict))
data_result['故障方式'] = data_new['故障情况'].apply(lambda x: identify_device(x, way_dict))
data_result['解决方案'] = data_new['处理情况'].apply(lambda x: identify_solution(x, sol_dict))
```

### 构建三元组

```python
triples = []
data_results_clean = data_result.drop(missing_rows)
for _, row in data_results_clean.iterrows():
    model, component, fault, solution = row

    triples.append((model, "车型", "包含部件", component, "部件"))
    triples.append((component, "部件", "故障类型", fault, "故障"))
    triples.append((fault, "故障", "解决方案", solution, "处理"))
    triples.append((component,'部件','解决方案', solution, '处理'))
```

```python
from py2neo import Graph, Node, Relationship, NodeMatcher

graph = Graph('neo4j://localhost:7687', auth=('neo4j', 'Wybxjhq2016'), name='neo4j')
matcher = NodeMatcher(graph)

for head, head_type, relation, tail, tail_type in triples:
    # 查找或创建头节点
    head_node = matcher.match(head_type, name=head).first()
    if not head_node:
        head_node = Node(head_type, name=head)
        graph.create(head_node)

    # 查找或创建尾节点
    tail_node = matcher.match(tail_type, name=tail).first()
    if not tail_node:
        tail_node = Node(tail_type, name=tail)
        graph.create(tail_node)

    # 创建关系
    rel = Relationship(head_node, relation, tail_node)
    graph.create(rel)
```
