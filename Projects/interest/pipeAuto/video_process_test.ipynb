```python
import cv2
import matplotlib.pyplot as plt
video = cv2.VideoCapture("data/test_video/2_2_chapter8_01.mp4")
```

```python
import os
frame_num = 400
save_step = 30
save_dir = "data/test_video/imgs"
os.makedirs(save_dir, exist_ok=True)
video.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
num = frame_num  # 从400开始计数
while True:
    ret, frame = video.read()
    if not ret:
        break
    if (num - frame_num) % save_step == 0:   # 每隔30帧保存
        cv2.imwrite(f"{save_dir}/{num}.jpg", frame)
        print(f"saved frame {num}")
    num += 1
```
