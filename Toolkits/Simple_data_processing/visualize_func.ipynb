```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

def plot_functions(f1, f2, name1 = 'func1', name2='func2', title='2 Functions Comparison', x_name='x', y_name='y', x_range=(-10, 10), num_points=1000):
    """
    绘制两个一元函数的图像
    
    参数:
    f1, f2: 要绘制的函数
    x_range: x轴范围，默认为(-10, 10)
    num_points: 采样点数量，默认为1000
    """
    # 创建x值数组
    x = np.linspace(x_range[0], x_range[1], num_points)
    
    # 计算对应的y值
    y1 = f1(x)
    y2 = f2(x)
    
    # 创建图形和坐标轴
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # 绘制函数图像
    ax.plot(x, y1, label=name1, color='blue', linewidth=2)
    ax.plot(x, y2, label=name2, color='red', linewidth=2)
    
    # 添加标题和标签
    ax.set_title(title, fontsize=16)
    ax.set_xlabel(x_name, fontsize=14)
    ax.set_ylabel(y_name, fontsize=14)
    
    # 添加网格
    ax.grid(True, linestyle='--', alpha=0.7)
    
    # 添加图例
    ax.legend(fontsize=12)
    
    # 显示图形
    plt.tight_layout()
    plt.show()
```

```python
# 定义两个函数
R=100
C=10e-6
L=470e-6

def function1(f):
    omega = 2*np.pi*f
    component1 = R*omega
    component2 = np.sqrt((1/C-L*omega**2)**2+(R*omega)**2)
    total = component1/component2
    return (total)

def function2(f):
    omega = 2*np.pi*f
    component1 = 1/C
    component2 = np.sqrt((1/C-L*omega**2)**2+(R*omega)**2)
    total = component1/component2
    return (total)

plot_functions(function1, function2, x_range=(50, 6000))
```
