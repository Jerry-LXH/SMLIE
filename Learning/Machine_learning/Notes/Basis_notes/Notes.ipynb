List and its methods:

```python
party_list=['tom',name,"mother"]
print(party_list)
party_list.sort(reverse=True) # 按字母反序排序
print("The sorted list is: "+str(party_list))
print("The first people is: "+party_list[0].title())
print("The last people is: "+party_list[-1].title())
party_list.append('dad') # 末尾添加
print("Add 'dad' and pop out the second: "+party_list.pop(1)) # pop方法返回弹出的元素
print("The remaining is: "+str(party_list))
print("Remove 'tom': ")
party_list.remove("tom") # romove方法无返回值
print("The remaining is: "+str(party_list))
print(name)
```

More about list:

```python
n=3 # power of x
number_list=[x**n for x in range(1,10)]
print(number_list)

copy_of_number_list=number_list[:] # 用slice切片复制数组
print(copy_of_number_list) # 数组更名只是创建了另一个引用对象的tag，必须用其他复制方式！

pair_of_list=(number_list,copy_of_number_list) # tuple 元组
print(pair_of_list)
for i in range(0,3): # for 循环
    pair_of_list[1].pop(i) # 正常的索引方法
print(pair_of_list) # 虽然元组本身形式不可变，但其中对象可变
```

Dictionary:

```python
party_list=['tom',name,"mother","dad"]
print(party_list)
number_list_of_party={}
for index,char in enumerate(party_list): # enumerate返回索引和值
    number_list_of_party[char]=index+1 # 为字典添加key-value pair
print(number_list_of_party)
del number_list_of_party['mother']
print("Now del 'mother' term: "+str(number_list_of_party))
for keys,items in number_list_of_party.items():
    print(keys+" is "+str(items)) #.key();.value()分别遍历
```

## <a id='toc1_3_'></a>[Class](#toc0_)

Model school courses with python class:

```python
class FudanCourse():
    def __init__(self,course_id,place) -> None:
        self.id=course_id
        self.place=place
        self.PNP=True # 自动初始化的属性
    def info(self):
        print("ID: "+self.id+"; Place: "+self.place+"; P/NP:"+str(self.PNP))
    def change_place(self,place):
        self.place=place

math_general=FudanCourse('14.2220','H3101')
math_general.info()
math_general.change_place('HGX303')
math_general.info()
```

Use inheritance to create Physics class:

```python
class Teacher:
    def __init__(self,department,name) -> None:
        self.department=department  
        self.name=name
    def change_name(self,name):
        self.name=name
    def change_department(self,dpt):
        self.department=dpt

class PhysicsCourse(FudanCourse):
    def __init__(self, course_id, place) -> None:
        super().__init__(course_id, place) # 让实例包含父类的所有属性
        self.experiment=True # 新的属性
        self.teacher=Teacher('Physics','?') # 用类的实例作为属性
    def info(self):
        print("ID: "+self.id+"; Place: "+self.place+"; P/NP:"+str(self.PNP)+
              "; Experiments:"+str(self.experiment)+"; Teacher: "+self.teacher.name+
              " from "+self.teacher.department)
physics_course_1=PhysicsCourse('14.2220','H3101')
physics_course_1.info()
physics_course_1.change_place('HGX303')
physics_course_1.teacher.change_name('Jackson') # 用多个连接号access Teacher类实例的属性和方法
physics_course_1.info()
```

类编码风格：
- 类名中的每个单词的**首字母**都大写,而不使用下划线
- 实例名和模块名都采用小写格式,并在单词之间加上下划线
- 类定义后面包含一个文档字符串：描述类的功能,遵循编写函数的文档时采用的格式约定
- 每个模块也都应包含一个文档字符串, 对其中的类可用于做什么进行描述
- 在类中,可使用一个空行来分隔**方法**;而在模块中, 可使用两个空行来分隔**类**
- 导入需要的模块时，先编写导入标准库模块的import语句,再添加一个空行, 然后编写导入你自己编写的模块的import语句

Test a class(whether methods work well):

```python
import unittest
from school import FudanCourse

class CourseTestCase(unittest.TestCase):

    """set up a class object"""
    def setUp(self) -> None:
        self.test_course=FudanCourse('14.22','H3109')
        self.changed_place='H3108'

    """Test the place change method:"""
    def test_place_changing_method(self):
        self.test_course.change_place(self.changed_place)
        self.assertEqual(self.test_course.place,self.changed_place)

suite = unittest.TestLoader().loadTestsFromTestCase(CourseTestCase)
unittest.TextTestRunner().run(suite)
```
