## python语言介绍
Python 是一种广泛使用的高级编程语言，由 Guido van Rossum 于1989年底发明，第一个公开发行版发行于1991年。Python 设计哲学强调代码的可读性和简洁的语法（尤其是使用空格缩进来区分代码块，而不是使用大括号或关键词）。这使得 Python 成为一种易于学习的语言，同时具备强大的功能，适合初学者和经验丰富的程序员。

以下是 Python 语言的一些关键特点：

**简洁易懂**：Python 的语法简洁，接近英语，使得代码易于阅读和编写。

**可移植性**：Python 代码可以在多种操作系统上运行，包括 Windows、Mac OS X、Linux 等。

**解释型语言**：Python 是一种解释型语言，这意味着开发过程中没有编译步骤，可以即时运行和测试代码。

**动态类型**：Python 允许在运行时改变变量类型，这使得编码更加灵活。

**自动内存管理**：Python 有自动垃圾回收机制，帮助管理内存使用。

**大型标准库**：Python 的标准库提供了大量的模块和函数，用于文件操作、系统调用、网络通信、文本解析、数学计算等。

**开源**：Python 是一个开源项目，由 Python 软件基金会管理，拥有活跃的社区支持。

**多范式编程**：Python 支持多种编程范式，包括面向对象、命令式、函数式和过程式编程。

**广泛的应用领域**：Python 在数据科学、机器学习、网络服务器、桌面应用程序、软件开发、自动化脚本等领域都有广泛应用。

丰富的框架和库：Python 拥有大量的第三方库和框架，如 Django（Web 开发）、Pandas（数据分析）、NumPy（科学计算）、TensorFlow 和 PyTorch（机器学习）等。
Python 的这些特性使其成为科学计算和数据分析的理想选择，同时也适合Web开发、自动化脚本、游戏开发等多种编程任务。Python 的语法和动态类型系统使得它非常适合快速开发和原型设计。此外，Python 的社区非常友好，为新入门的程序员提供了大量的学习资源和支持。

## 1.2 容器(list与dict)  
### 1.2.1 list  
list是一种有序的集合，可以随时添加和删除其中的元素。list用方括号[]表示，元素之间用逗号分隔。     
**掌握list的基本操作**   
创建list：my_list = [1, 2, 3]  
访问元素：my_list[0]（获取第一个元素）  
修改元素：my_list[1] = 4（将第二个元素修改为4）  
添加元素：my_list.append(5)（在列表末尾添加元素5）  
删除元素：my_list.remove(3)（删除元素3）  
列表切片：my_list[1:3]（获取索引1到2的元素）  
列表长度：len(my_list)  

### 1.2.2 dict
dict是一种键值对（key-value pair）的集合，其中的键是唯一的。dict用大括号{}表示，键值对之间用逗号分隔。  
  
**掌握dict的基本操作**  
创建dict：my_dict = {'a': 1, 'b': 2, 'c': 3}  
访问元素：my_dict['a']（获取键'a'对应的值）  
修改元素：my_dict['b'] = 4（将键'b'对应的值修改为4）  
添加元素：my_dict['d'] = 5（添加一个新的键值对）  
删除元素：del my_dict['c']（删除键为'c'的键值对）  
检查键是否存在：'a' in my_dict（返回True或False）  
获取所有键：my_dict.keys()  
获取所有值：my_dict.values()  
获取所有键值对：my_dict.items()获取所有键值对：my_dict.items()

# 2 控制
## 2.1 循环控制  
循环的写法有while循环，for循环。   
```python
i = 5
while i > 0:
    i = i - 1
    print(i)

a = [1, 2, 3, 4, 5]
for i in a:
    print(i)
    if i >= 3:
        break
```
下面使用的for循环，结果4950是因为range函数从start开时但不到stop。 range(start, stop[, step])

````python
v = 0
for i in range(1, 100):
    v = v + i
v
````

## 2.2 条件控制

```python
if condition1:
# 如果condition1为真，则执行这里的代码
elif condition2:
# 如果condition1为假且condition2为真，则执行这里的代码
else:
# 如果上述条件都不为真，则执行这里的代码不```的代码
```

````python
flag = False
name = 'luren'
if name == 'python':         # 判断变量是否为 python 
    flag = True              # 条件成立时设置标志为真
    print('welcome boss')     # 并输出欢迎信息
else:
    print(name)               # 条件不成立时输出变量名称
````
