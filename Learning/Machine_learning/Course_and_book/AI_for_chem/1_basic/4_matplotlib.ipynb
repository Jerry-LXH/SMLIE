```python
# 导入matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()             # 创建一个包含一个axes的figure
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # 在axes上绘制折线图
plt.show()                           # 显示图像
```

```python
# 绘制多个图像
x = np.linspace(0, 2, 100)            # 生成0到2之间的100个等差数
y = np.sin(2 * np.pi * x)             # 计算正弦函数值
fig, axes = plt.subplots(2)          # 创建一个包含2个axes的figure
axes[0].plot(x, y)                   # 在第一个axes上绘制折线图
axes[1].scatter(x, y)                # 在第二个axes上绘制散点图
plt.show()                           # 显示图像
```

以下代码示例来源于[官网样例Power spectral density (PSD)](https://matplotlib.org/stable/gallery/lines_bars_and_markers/psd_demo.html#sphx-glr-gallery-lines-bars-and-markers-psd-demo-py)

```python
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.mlab as mlab

# Fixing random state for reproducibility
np.random.seed(19680801)

dt = 0.01
t = np.arange(0, 10, dt)
nse = np.random.randn(len(t))
r = np.exp(-t / 0.05)

cnse = np.convolve(nse, r) * dt
cnse = cnse[:len(t)]
s = 0.1 * np.sin(2 * np.pi * t) + cnse

fig, (ax0, ax1) = plt.subplots(2, 1, layout='constrained')
ax0.plot(t, s)
ax0.set_xlabel('Time (s)')
ax0.set_ylabel('Signal')
ax1.psd(s, 512, 1 / dt)

plt.show()
```
